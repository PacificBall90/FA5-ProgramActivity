```python
def price(cost, quantity):
  total = cost * quantity
  return total

cost = int(input("Input price of product: "))
quantity = int(input("Input number of products: "))
print()
if cost and quantity:
 print("The total cost is: ₱", price(cost, quantity))
else:
  print("ERROR: Invalid data")
```
